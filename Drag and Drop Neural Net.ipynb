```python
# given list of layer specifications, build the neural net
def buildNet(layerList):
    # currently only allow Sequential
    netAsString = "from keras.models import Sequential\n"
    
    # add imports with header dictionary
    headerDict = layerList[0]
    for imp in headerDict['imports']:
        netAsString += imp + "\n"
    
    # for prettier spacing
    netAsString += "\n"
    
    # init model
    netAsString += 'model = Sequential()\n\n'
    
    # build each layer
    for i in range(1, len(layerList) - 1):
        netAsString += createLayer(layerList[i])
    
    # compile the model with footer dictionary
    footerDict = layerList[-1]
    netAsString += "model.compile("
    
    for spec in footerDict:
        netAsString += spec + "=" + footerDict[spec] + ","
        
    # remove comma TODO: find cleaner way to do this
    netAsString = netAsString[:-1]
    netAsString += ")\n"
    
    print(netAsString)

    
    
# calls relevant layer creation function
def createLayer(layerDict):
    if layerDict['layerType'] == 'Dense':
        return createDense(layerDict)
    elif layerDict['layerType'] == 'Activation':
        return createActivation(layerDict)
    elif layerDict['layerType'] == 'Dropout':
        return createDropout(layerDict)
    

    
# creates Dense layer to the specifications of dictionary input
def createDense(denseDict):
    # display number of units first
    layerAsString = "model.add(Dense(units=" + str(denseDict['units'])
    
    # add additional specifications
    for key in denseDict:
        if key != 'units' and key != 'layerType':
            layerAsString += "," + key + "=" + str(denseDict[key])
            
    layerAsString += "))\n"
    
    # print dense layer
    return layerAsString
    
    
    
# creates Activation layer to the specifications of dictionary input
def createActivation(activationDict):
    # only requires activation input
    # TODO: allow Theano or TensorFlow op
    layerAsString = "model.add(Activation(activation=" + str(activationDict['activation']) + "))\n"
    
    # print dense layer
    return layerAsString
    
    
    
# creates Dropout layer to the specifications of dictionary input
def createDropout(dropoutDict):
    # display rate first
    layerAsString = "model.add(Dropout(rate=" + str(dropoutDict['rate'])
    
    # add additional specifications
    for key in dropoutDict:
        if key != 'rate' and key != 'layerType':
            layerAsString += "," + key + "=" + str(dropoutDict[key])
            
    layerAsString += "))\n"
    
    # print dense layer
    return layerAsString
```

```python
net = []

header = {}
header['imports'] = ['from keras.layers import Dense, Dropout, Activation']
net.append(header)

dense = {}
dense['layerType'] = 'Dense'
dense['units'] = 32
dense['activation'] = "'relu'"
dense['use_bias'] = False
net.append(dense)

dropout = {}
dropout['layerType'] = 'Dropout'
dropout['rate'] = .2
dropout['seed'] = 7
net.append(dropout)

activation = {}
activation['layerType'] = 'Activation'
activation['activation'] = "'relu'"
net.append(activation)

footer = {}
footer['optimizer'] = "'rmsprop'"
footer['loss'] = "'categorical_crossentropy'"
footer['metrics'] = "['accuracy']"
net.append(footer)
```

```python
buildNet(net)
```
